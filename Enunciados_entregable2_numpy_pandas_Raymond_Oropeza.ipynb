### 🎈 <font color='red'> 1. Ejercicio ventas 🎈

Tenemos un dataset de ventas de 7 días con 3 productos diferentes. Queremos:

1. Convertir la lista en un array 2D.

2. Calcular la venta total por producto y la venta total por día.

3. Encuentrar el día con mayor venta total y el producto más vendido.

```python
# dataset de ventas
ventas = [
    [5, 7, 3],
    [8, 6, 5],
    [4, 5, 6],
    [7, 8, 4],
    [6, 7, 8],
    [5, 6, 7],
    [9, 8, 6]
]
```

```python
import numpy as np
#Convertir la lista en un array 2D.
array2d= np.array(ventas)
print(array2d)
print(type(array2d))
```

```python
# Calcular la venta total por producto y la venta total por día.

venta_total_producto = sum(array2d)
print(venta_total_producto)

ventas_por_dia = array2d.sum(axis=1)
print(ventas_por_dia)
```

```python
# Encuentrar el día con mayor venta total y el producto más vendido.
dia_max_ventas= max(ventas_por_dia)
print(dia_max_ventas)

producto_mas_vendido= max(venta_total_producto)
print(producto_mas_vendido)
```

### 🎈 <font color='red'> 2. Ejercicio datos financieros 🎈

Tenemos un dataset de 10 clientes con sus ingresos y gastos mensuales. Queremos:

1. Convierte la lista en un array 2D.

2. Crea una nueva columna que sea el ahorro mensual (ingresos - gastos).

3. Filtra los clientes con ahorro superior a 1000 y calcula su ahorro promedio.

```python
# dataset

datos = [
    [2500, 1500],
    [3000, 2000],
    [1800, 1000],
    [4000, 3500],
    [2200, 1800],
    [3500, 1500],
    [2800, 1200],
    [3200, 2500],
    [2700, 2700],
    [3600, 3000]
]
```

```python
#Convierte la lista en un array 2D.

array2d = np.array(datos)
array2d
```

```python
#Crea una nueva columna que sea el ahorro mensual (ingresos - gastos).
import pandas as pd
import numpy as np

df = pd.DataFrame(array2d, columns=['ingresos','gastos'])
df

df['ahorro_al_mes']=df['ingresos']- df['gastos']
df
```

```python
#Filtra los clientes con ahorro superior a 1000 y calcula su ahorro promedio.
df_filtrado = df[ df["ahorro_al_mes"] >1000]
print(df_filtrado)

promedio_ahorro = df_filtrado['ahorro_al_mes'].mean()

print(df_filtrado)
print("Ahorro promedio:", promedio_ahorro)
```

### 🎈 <font color='red'> 3. Ejercicio calificaciones de estudiantes 🎈

Tenemos una dataset de calificaciones de 5 estudiantes en 4 asignaturas y los nombres de los 5 estudiantes. Queremos:

1. Convertir la lista en un array 2D de NumPy.

2. Calcular la media por estudiante y la media por asignatura.

3. Encuentrar los estudiantes (nombres) con promedio superior a 8.

```python
import pandas as pd
import numpy as np
```

```python
# Nombres de los estudiantes
nombres = ["Ana", "Luis", "María", "Carlos", "Elena"]

# calificaciones

notas = [
    [8, 7, 9, 10],
    [6, 5, 7, 8],
    [9, 9, 8, 10],
    [7, 6, 7, 6],
    [10, 10, 9, 10]
]
```

```python
#Convertir la lista en un array 2D de NumPy

array2d = np.array(notas)
array2d
```

```python
#Calcular la media por estudiante y la media por asignatura.

df = pd.DataFrame(
    array2d,
    index=nombres,
    columns=['asignatura1','asignatura2','asignatura3','asignatura4']
)
print(df)

media_por_estudiante = df.mean(axis=1)
media_por_estudiante

media_por_asignatura= df.mean(axis=0)
media_por_asignatura
```

```python
#Encuentrar los estudiantes (nombres) con promedio superior a 8.

nombres_filtrados = df.index[media_por_estudiante > 8]
nombres_filtrados
```
